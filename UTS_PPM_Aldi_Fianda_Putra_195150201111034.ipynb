### 6.1 K-Nearest Neighboor dan Normalisasi Data

Sebelum kita menghitung K-Nearest Neighboor ataupun metode algoritma yang lainnya. Kita dapat melakukan pemrosesan data awal dulu agar nantinya kita dapat lebih mudah dalam menghitung datanya, dalam hal ini yang dapat dilakukan adalah melakukan normalisasi yang mana dapat memudahkan proses penghitungan baik untuk algoritma KNN, Naive Bayes ataupun Decision Tree. Untuk itu blok kode program berikut berfungsi untuk menampilkan rentang fitur dari dataset Dry Bean kecuali fitur terakhirnya, karena fitur terakhir merupakan fitur yang bersifat kualitatif sehingga tidak dapat dihitung. Blok kode program berikut berfungsi untuk menghitung rentang fitur dari semua data.

```python
def rentang(input):
  print('Berikut adalah rentang dari masing-masing fitur :')
  daftar_fitur=input.columns[:-1] # Karena bagian terakhir adalah kelas yang mana berupa string sehingga tidak bisa dihitung dengan numerik
  for fitur in daftar_fitur:
    max=input[fitur].max()
    min=input[fitur].min()
    print("Rentang fitur", fitur, "Adalah", max-min)
```

Karena permintaan pada soal adalah dengan menggunakan Euclidean, maka berikut kodingan untuk rumus menghitung jarak euclidean beserta rumus KNN. Selain itu karena pada penghitungan KNN sendiri hanya dapat menghitung fitur yang bersifat numerikm, maka kita dapat memisahkan label yang mana pada kasus ini adalah fitur Kelas dengan data latih dan data uji dengan cara melakukan pop terhadap datanya yakni seperti blok kode program berikut.

```python
label_latih = data_latih.pop('Class')
label_uji = data_uji.pop('Class')
```

```python
def euclidean(input_1,input_2):
    return np.sqrt(np.sum(np.square(input_1-input_2)))
```

```python
# rumus KNN
from collections import Counter
def knn(k,data_latih,label,data_uji):
    jarak = np.array([euclidean(data_latih.iloc[x],data_uji) for x in range(data_latih.shape[0])])
    min_k = jarak.argsort()[:k]
    k_kelas = label.iloc[min_k].to_numpy()
    hitung = Counter(k_kelas)
    kelas_terdekat = hitung.most_common(1)[0][0]
    return kelas_terdekat
```

Proses berikut akan memakan waktu lumayan lama karena menguji 14 data dengan setiap satu data uji diujikan dengan data latih yang jumlahnya 13597 buah.

```python
def hitung_prior(kelas):
  i=0
  prior = Counter(kelas)
  n_kelas = prior.keys()
  n_data = kelas.shape[0]
  for i in n_kelas:
    prior[i]=prior[i]/n_data
  return prior
```

```python
prior = hitung_prior(data_latih['Class'])
prior
```

```python
def rata2_std_kelas(df_input):
  kolom = df_input.columns[:-1]
  kelas = df_input.columns[-1]
  daftar_kelas = set(df_input[kelas])
  rata2 = {}
  std = {}
  for k in kolom:
    for x in daftar_kelas:
      rata2[(x,k)] = df_input.loc[df_input[kelas]==x][k].mean()
      std[(x,k)] = df_input.loc[df_input[kelas]==x][k].std()
  return(rata2,std)
```

```python
def likelihood_gaussian(df_input,rata2,std):
  rst = 1/np.sqrt(2*math.pi*(np.square(std)))
  rst = rst*np.exp((-1*(np.square(df_input-rata2)))/(2*np.square(std)))
  return rst
```

```python
def train_bayes_gaussian(data_latih):
  kelas = data_latih.columns[-1]
  prior = hitung_prior(data_latih[kelas])
  (rata2,std) = rata2_std_kelas(data_latih)
  daftar_kelas = set(data_latih[kelas])
  daftar_kolom = data_latih.columns[:-1]
  model = {}
  model['prior'] = prior
  model['rata2'] = rata2
  model['std'] = std
  model['daftar_kelas'] = daftar_kelas
  model['daftar_kolom'] = daftar_kolom
  return model

model = train_bayes_gaussian(data_latih)
```

```python
def testing_bayes_gaussian(model,data_uji):
  prior = model['prior']
  rata2 = model['rata2']
  std = model['std']
  daftar_kelas = model['daftar_kelas']
  kolom = model['daftar_kolom']
  posterior = dict.fromkeys(daftar_kelas,1)
  for x in daftar_kelas:
    for k in kolom:
      posterior[x] = posterior[x]*likelihood_gaussian(data_uji[k],rata2[(x,k)],std[(x,k)])
    posterior[x] = posterior[x]*prior[x]
    print('Nilai dari posterior terhadap kelas',x,'adalah :', posterior[x])
  kelas_uji = max(posterior,key=posterior.get)
  return kelas_uji
```

```python
def uji_posterior(index):
  print('Jenis Dry Bean yang diuji adalah     :',label_uji.iloc[index])
  print("Index ke   :",index)
  print('Atribut    : ')
  print(data_uji.iloc[index])
  print()
  prediksi = testing_bayes_gaussian(model,data_uji.iloc[index])
  print()
  print('Maka       : ')
  print('Jenis Dry Bean cocok berdasarkan metode Naive Bayes adalah :',prediksi)
```
